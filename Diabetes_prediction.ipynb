```python
dataframe['Pregnancies'] = dataframe['Pregnancies'].replace(0, dataframe['Pregnancies'].median())
dataframe['Glucose'] = dataframe['Glucose'].replace(0, dataframe['Glucose'].mean())
dataframe['BloodPressure'] = dataframe['BloodPressure'].replace(0, dataframe['BloodPressure'].mean())
dataframe['SkinThickness'] = dataframe['SkinThickness'].replace(0, dataframe['SkinThickness'].median())
dataframe['BMI'] = dataframe['BMI'].replace(0, dataframe['BMI'].mean())
dataframe['DiabetesPedigreeFunction'] = dataframe['DiabetesPedigreeFunction'].replace(0, dataframe['DiabetesPedigreeFunction'].median())
dataframe['Age'] = dataframe['Age'].replace(0, dataframe['Age'].median())
```

EDA-OUTLIERS DETECTION AND NORMALIZATIONS

```python
dataframe.columns
```

```python
#SPLIT THE DATASET
#x=Features
#y = Target
X=dataframe.drop(("Outcome"),axis=1)
Y=dataframe["Outcome"]
```

```python
fig,ax=plt.subplots(figsize=(15,15))
sns.boxplot(data=X)
```

```python
cols = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']

for col in cols:
    Q1 = X[col].quantile(0.25)
    Q3 = X[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    mask = (X[col] >= lower_bound) & (X[col] <= upper_bound)
```

```python
x_outlier_detection= X[mask]
y_outlier_detection= Y[mask]
```

```python
X[mask].shape,Y[mask].shape,x_outlier_detection.shape
```

Normalization

```python
from sklearn.preprocessing import StandardScaler
standard=  StandardScaler()
x_normalized=standard.fit_transform(x_outlier_detection)
```

```python
fig,ax=plt.subplots(figsize=(15,15))
sns.boxplot(x_normalized)
```

```python
x_normalized_df=pd.DataFrame(x_normalized,columns=X.columns)
```

```python
fig,ax=plt.subplots(figsize=(15,15))
sns.boxplot(x_normalized_df)
```

```python
q= x_normalized_df["Insulin"].quantile(.95)
mask = x_normalized_df["Insulin"] < q
x_normalized_df = x_normalized_df.reset_index(drop=True)
y_outlier_detection = y_outlier_detection.reset_index(drop=True)
datanew = x_normalized_df[mask]
y_outlier_detection = y_outlier_detection[mask]

```

MODEL TRAINING

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(datanew, y_outlier_detection, test_size=0.33, random_state=42)
```

```python
X_train.shape,X_test.shape
```

IMBLANCING OF DATA(undersampling,oversampling)


*   oversampling
   * minorty class and increse that number to the majority class

*  Undersampling
   * majority class and decrrease the number of minority class
* smote
  * create synthetic data and increse the no of sample

* imb class



```python
y_train.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
x_train_resampled,y_train_resampled=smote.fit_resample(X_train,y_train)
```

```python
y_train.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
classification =LogisticRegression()
classification.fit(x_train_resampled,y_train_resampled)
```

model prediction

```python
y_pred=classification.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
target_names = ['Dibetic', 'Non Dibetic']
print(classification_report(y_test,y_pred, target_names=target_names))
```
